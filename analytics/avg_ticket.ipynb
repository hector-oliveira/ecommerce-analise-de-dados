```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/customer_clear.csv")
```

```python
# Calcula o ticket médio para compras com e sem desconto
# (permite avaliar se o desconto aumenta ou reduz o valor gasto)
mean_discount = df.groupby('discount_applied')['purchase_amount_usd'].mean()
display(mean_discount)
```

```python
# Ajusta os rótulos para facilitar leitura dos resultados no gráfico
labels = ['Sem desconto', 'Com desconto']
```

```python
# Define cores: cinza para compras normais e verde para compras com desconto
colors = ['grey', 'green']
```

```python
# Visualiza o ticket médio comparando compras com e sem desconto
plt.figure(figsize=(6,4))
mean_discount.plot(kind='bar', color=colors)
plt.title("Ticket médio: com desconto vs. sem desconto")
plt.xlabel("Aplicação de desconto")
plt.ylabel("Valor médio da compra (USD)")
plt.xticks([0,1], labels, rotation=0)
plt.show()
```

# 3.1 — Cupom e Desconto Aumentam o Valor Gasto?
### Análise do impacto do desconto aplicado no valor gasto

## Objetivo  
Avaliar se a aplicação de desconto aumenta ou reduz o valor médio das compras.  
Essa análise permite responder:

> O cliente realmente gasta mais quando recebe desconto?

---

## Resultado Observado

| Condição         | Ticket Médio (USD) |
|------------------|--------------------|
| Sem desconto     | ~60,13             |
| Com desconto     | ~59,28             |

O valor médio pago por clientes que recebem desconto é menor do que o valor pago por clientes sem desconto.

---

## Insight  
A concessão de desconto não aumenta o valor gasto.  
Clientes que recebem desconto tendem a gastar menos, em vez de ampliar o valor da compra. Isso indica que o desconto pode estar atraindo consumidores que:

- compram somente quando há benefício;
- possuem baixo gasto natural;
- não ampliam o valor do carrinho quando recebem desconto.

---

## Recomendação Estratégica  
A aplicação de desconto deve ser seletiva, e não geral. Boas práticas incluem:

- uso de desconto apenas para queima de estoque;
- aplicação em produtos com baixa demanda;
- utilização em campanhas sazonais específicas;
- desconto condicionado a valor mínimo de compra  
  *(ex.: oferecer desconto somente para compras a partir de um valor estabelecido).*

Descontos condicionados podem estimular aumento do ticket médio, ao contrário do desconto indiscriminado, que tende a reduzi-lo.
