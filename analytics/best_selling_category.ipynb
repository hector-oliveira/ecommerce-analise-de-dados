```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar tabela limpa
df = pd.read_csv("../data/customer_clear.csv")
```

```python
# Quantifica o volume de vendas por categoria para entender quais áreas do negócio geram mais receita

category_counts = df['category'].value_counts()
```

```python
# Gráfico para comparar visualmente o desempenho de cada categoria

plt.figure(figsize=(8,4))
category_counts.plot(kind='bar')
plt.title("Categorias mais vendidas")
plt.xlabel("Categoria")
plt.ylabel("Quantidade de Vendas")
plt.xticks(rotation=45)
plt.show()
```

```python
# Identifica os itens mais populares (top 10) para destacar produtos com maior potencial comercial
item_counts = df['item_purchased'].value_counts().head(10)
```

```python
# Visualização dos itens líderes de venda — útil para decisões de marketing, precificação e estoque

plt.figure(figsize=(10,4))
item_counts.plot(kind='bar')
plt.title("Top 10 Itens mais vendidos")
plt.xlabel("Itens")
plt.ylabel("Quantidade de Vendas")
plt.xticks(rotation=45)
plt.show()
```

# Insights esperados dessa análise

## 1) Categorias mais vendidas

Ao observar o gráfico de categorias (quantidade de vendas), conseguimos:

- Identificar quais categorias são o **núcleo do negócio** (geralmente vestuário no e-commerce).
- Encontrar categorias com **alta demanda natural**, que não precisam de promoções fortes.
- Detectar categorias com **baixo volume**, que podem:
  - necessitar de promoção
  - ser descontinuadas
  - ser reposicionadas

### Insight estratégico
**Se roupas forem os produtos mais vendidos, o marketing deve focar nesse segmento, com maior variedade e reposição rápida.**

---

## 2) Top 10 itens mais vendidos

Ao analisar os itens individualmente, é possível:

- Identificar os **produtos líderes de venda**.
- Criar **kits promocionais** usando itens populares.
- **Aumentar a margem** ajustando preço de produtos com alta procura sem comprometer o volume de vendas.
- Sugerir automaticamente **itens complementares (cross-sell)**.

### Insight estratégico
**Produtos mais vendidos podem impulsionar a venda de acessórios ou itens de ticket maior.**
