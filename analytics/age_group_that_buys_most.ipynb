```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/customer_clear.csv")
```

```python
# Cria faixas etárias para segmentar o comportamento de compra por idade

bins = [0, 18, 25, 35, 50, 65, 120]
labels = ['menor que 18', '18-25', '26-35', '36-50', '51-65', '65+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
```

```python
# Contabiliza o número de compradores por faixa etária para entender público dominante

age_counts = df['age_group'].value_counts().sort_index()
display(age_counts)
```

```python
# Visualiza a distribuição de compras entre grupos de idade

plt.figure(figsize=(7,4))
age_counts.plot(kind='bar')
plt.title("Faixa etária que mais compra")
plt.xlabel("Faixa etária")
plt.ylabel("Quantidade de compras")
plt.show()
```

```python
# Avalia a participação de gênero nas compras, indica foco de marketing e mix de produtos

gender_counts = df['gender'].value_counts()
display(gender_counts)
```

```python
# Visualiza a representatividade de vendas entre homens e mulheres

plt.figure(figsize=(6,4))
gender_counts.plot(kind='bar')
plt.title("Compras por gênero")
plt.xlabel("Gênero")
plt.ylabel("Quantidade de compras")
plt.show()
```

```python
# Identifica as principais regiões consumidoras (top 10) para direcionar ações geográficas

location_counts = df['location'].value_counts().head(10)

display(location_counts)
```

```python
# Visualiza as localizações com maior volume de compras

plt.figure(figsize=(10,4))
location_counts.plot(kind='bar')
plt.title("Top 10 localizações com mais compras")
plt.xlabel("Localização")
plt.ylabel("Quantidade de compras")
plt.xticks(rotation=45)
plt.show()
```

# Insights – Perfil de Quem Mais Compra

## 1) Faixa etária que mais compra

### Distribuição de clientes por faixa de idade
| Faixa etária | Quantidade |
|--------------|------------|
| < 18         | 0          |
| 18–25        | 486        |
| 26–35        | 755        |
| 36–50        | 1100       |
| 51–65        | 1132       |
| 65+          | 427        |

### Interpretação
O maior volume de compras se concentra entre **36 e 65 anos**, com pico entre **51–65 anos**. Trata-se de um público mais maduro, com maior poder de compra e menor comportamento impulsivo.

### Implicações práticas
- Estratégias devem priorizar **conforto, utilidade e qualidade**, ao invés de tendências rápidas.
- Comunicação deve destacar benefícios **durabilidade, praticidade e custo-benefício**.
- Produtos premium podem ter mais adesão entre **36–65 anos**, sem depender de descontos agressivos.

### Recomendação estratégica
**Posicionar campanhas e mix de produtos com foco em necessidades práticas e valor percebido para clientes entre 36 e 65 anos.**
